# Import required library
# Load Dataset

```python
import pandas as pd
df=pd.read_csv('wine.csv')
```

# Analyze your dataset
# Look for missing values
# Look for outliers
# Verify if any pre-processing is required

```python
print(df.info())#For datatypes and Null count
print(df.describe(include='all')) # For statistics
```

```python
uclass=df['class'].unique()
print(uclass)
```

# Store features in X
# Store Labels/Class in Y
# Split Dataset into Training and Test
# To validate Results

```python
from sklearn.model_selection import train_test_split
Y=df['class']
X=df.drop('class',axis=1)
print(X.head())
print(Y.head())
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

# No missing Values
# Variability in columns
# Normailize all columns
# So that all columns have equal contribution in learning
# Use Scikit Library for normalizing

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler=StandardScaler()
Xs_train=scaler.fit_transform(X_train)#Normalizes Training Data
Xs_test=scaler.fit_transform(X_test)#Normalizes Test Data
print(np.mean(Xs_test,axis=0))
print(np.var(Xs_test,axis=0))
```

# Now Train your model

```python
from sklearn.neighbors import KNeighborsClassifier
myclassifier=KNeighborsClassifier(n_neighbors=4)# Define your model
myclassifier.fit(Xs_train,Y_train)#Train your model
```

# Now do some predictions on Test Data

```python
Y_pred=myclassifier.predict(Xs_test)#Makes class prediction for each sample
print(Y_pred)
```

# Check with metrics if your algorithm is working fine

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(f"Accuracy:{accuracy_score(Y_test,Y_pred)}")
print(f"Classifiaction Report:{classification_report(Y_test,Y_pred)}")
print(f"confusion_matrix:{confusion_matrix(Y_test,Y_pred)}")
```
