#Import Necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#For specific model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error# For regression
from sklearn.metrics import accuracy_score,confusion_matrix#For classification
```

```python
df=pd.read_excel('realestatedataset.xlsx')
print(df.info())
```

```python
X=df[['X6 longitude','X2 house age']]
Y=df['Y house price of unit area']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=48)
```

```python
s=StandardScaler()
Xs_train=s.fit_transform(X_train)
Xs_test=s.fit_transform(X_test)
```

```python
mymodel=LinearRegression()
mymodel.fit(Xs_train,Y_train)
ypred=mymodel.predict(Xs_test)
print(list(Y_test))
print(ypred)
print(mymodel.coef_,mymodel.intercept_)
```

#Metrics

```python
print(f"Mean Error:{mean_absolute_error(Y_test,ypred)}")
plt.plot(list(Y_test),label='Actual',color='blue')
plt.plot(ypred,label='predicted',color='red')
plt.legend()
plt.show()
```

```python
Yp_train=mymodel.predict(Xs_train)
print(mean_absolute_error(Y_train,Yp_train))
```
